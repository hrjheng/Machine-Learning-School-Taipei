```python
import ROOT
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from skhep.visual import MplPlotter as skh_plt
import os
import tdrstyle as tdr

tdr.setTDRStyle()
ROOT.gStyle.SetOptStat(0)
# ROOT.gStyle.SetOptFit(0)

%jsroot off
```

```python
f_hist = ROOT.TFile('./TMVARegApp_TestSample_TrainSampleCentMerged.root','READ')
hM_pred_ratio_BDT = ROOT.gROOT.FindObject('hM_pred_ratio_BDT')
hM_pred_ratio_BDTG = ROOT.gROOT.FindObject('hM_pred_ratio_BDTG')
hM_pred_ratio_DNN_CPU = ROOT.gROOT.FindObject('hM_pred_ratio_DNN_CPU')
hM_pred_ratio_MLP = ROOT.gROOT.FindObject('hM_pred_ratio_MLP')
list_hist = []
list_hist.append(hM_pred_ratio_BDT)
list_hist.append(hM_pred_ratio_BDTG)
list_hist.append(hM_pred_ratio_DNN_CPU)
list_hist.append(hM_pred_ratio_MLP)

hM_pred_ratio_BDT.Scale(1./hM_pred_ratio_BDT.Integral(-1,-1))
hM_pred_ratio_BDTG.Scale(1./hM_pred_ratio_BDTG.Integral(-1,-1))
hM_pred_ratio_DNN_CPU.Scale(1./hM_pred_ratio_DNN_CPU.Integral(-1,-1))
hM_pred_ratio_MLP.Scale(1./hM_pred_ratio_MLP.Integral(-1,-1))

list_color = ['#ff5959','#ffad5a','#4f9da6','#1a0841']
list_method = ['BDT','BDTG','DNN','MLP']
```

```python
def Gaussian_fit(hist,colorname):
    func = ROOT.TF1('func','gaus',0,2)
    func.SetLineColor(ROOT.TColor.GetColor(colorname))
    func.SetLineWidth(3)
    ROOT.gStyle.SetOptFit(0);
    ROOT.gStyle.SetOptStat(0);
    hist.Fit(func)
    return func, (func.GetParameter(1),func.GetParError(1)), (func.GetParameter(2),func.GetParError(2))
```

```python
c0 = ROOT.TCanvas('c0', 'c0', 800, 700)
c0.cd()
c0.Update()
ROOT.gPad.SetLeftMargin(0.15)
ROOT.gPad.SetBottomMargin(0.16)
ROOT.gPad.SetRightMargin(0.06)
leg = ROOT.TLegend(0.2, 0.65, 0.93, 0.9)

for i in range(len(list_hist)):
    list_hist[i].SetLineWidth(3)
    list_hist[i].SetLineColor(ROOT.TColor.GetColor(list_color[i]))
    if i == 0:
        list_hist[i].GetXaxis().SetTitle('(#sum p_{T}^{jet region})/prediction')
        list_hist[i].GetYaxis().SetTitle('Normalized entries')
        list_hist[i].GetXaxis().SetRangeUser(0,3)
        list_hist[i].GetYaxis().SetRangeUser(0, 0.3)
        list_hist[i].GetXaxis().SetTitleOffset(1.3)
        list_hist[i].GetYaxis().SetTitleOffset(1.5)
        list_hist[i].GetXaxis().SetTitleSize(0.05)
        list_hist[i].GetYaxis().SetTitleSize(0.05)
        list_hist[i].GetXaxis().SetLabelSize(0.05)
        list_hist[i].GetYaxis().SetLabelSize(0.05)
        list_hist[i].Draw('hist')
    else:
        list_hist[i].Draw('histsame')
        
    fit, (mean,meanerr), (sigma,sigmaerr) = Gaussian_fit(list_hist[i],list_color[i])
    fit.Draw('lsame')
        
    leg.AddEntry(list_hist[i], list_method[i]+' (#mu={0:.3f}#pm{1:.2e}; #sigma={2:.3f}#pm{3:.2e})'.format(mean, meanerr, sigma, sigmaerr), 'l')

leg.SetBorderSize(0)
leg.SetFillStyle(0)
leg.Draw('same')

c0.Draw()
c0.SaveAs('./plots/Training-Result-TrainSampleCentMerged-TestSampleCent0to30.png')
c0.SaveAs('./plots/Training-Result-TrainSampleCentMerged-TestSampleCent0to30.pdf')
```
