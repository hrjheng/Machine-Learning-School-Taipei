```python
import ROOT
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from skhep.visual import MplPlotter as skh_plt
import os
import tdrstyle as tdr

tdr.setTDRStyle()
ROOT.gStyle.SetOptStat(0)
ROOT.gStyle.SetOptFit(0)

%jsroot off
```

```python
f = ROOT.TFile('./data/MinBias_Hydjet_Drum5F_2018_5p02TeV_minPt0p1_cent0to100.root','READ')
tree = ROOT.gROOT.FindObject("EventTree")
```

```python
hM = ROOT.TH2F('hM','hM',100,0,3000,100,0,40000)
hM_x = hM.ProjectionX()

profile_cp = ROOT.TProfile()

nEntries =  tree.GetEntries()
for iEntry in range(0,nEntries):
    tree.GetEntry(iEntry)
    hM.Fill(tree.Ncoll,tree.Nparticle)
    

c = ROOT.TCanvas('c', 'c', 1600, 700)
c.Divide(2,1)
c.cd(1).SetLogz()
ROOT.gStyle.SetPalette(57)
# ROOT.TColor.InvertPalette()
ROOT.gPad.SetRightMargin(0.13)
ROOT.gPad.SetLeftMargin(0.16)
ROOT.gPad.SetBottomMargin(0.16)
hM.GetXaxis().SetTitle('Number of collisions')
hM.GetYaxis().SetTitle('Number of particles')
hM.GetYaxis().SetLabelSize(0.04)
hM.GetXaxis().SetTitleOffset(1.1)
hM.GetYaxis().SetTitleOffset(1.4)
hM.SetContour(2000)
hM.Draw('COLZ')

profile = hM.ProfileX()
profile.Draw('sameE1')

func1 = ROOT.TF1('func1','pol4',0,2000)
func1.SetLineColor(2)
func1.SetLineWidth(2)
profile.Fit(func1,'','',0,2000)
profile.Copy(profile_cp)

ROOT.gROOT.ForceStyle()

c.cd(2)
ROOT.gPad.SetRightMargin(0.05)
ROOT.gPad.SetBottomMargin(0.16)
for ibinx in range(profile_cp.GetNbinsX()):
    hM_x.SetBinContent(ibinx+1,profile.GetBinContent(ibinx+1)/func1.Eval(profile.GetBinCenter(ibinx+1)))
    
hM_x.GetXaxis().SetTitle('Number of collisions')
hM_x.GetYaxis().SetTitle('Profile/Function')
hM_x.GetYaxis().SetRangeUser(0.8,1.2)
hM_x.GetYaxis().SetLabelSize(0.05)
hM_x.GetXaxis().SetTitleOffset(1.1)
hM_x.GetYaxis().SetTitleOffset(1.4)
hM_x.SetMarkerSize(1.5)
hM_x.Draw('p')

c.Draw()
c.SaveAs('./plots/Exercise_Monday_1.png')
c.SaveAs('./plots/Exercise_Monday_1.pdf')
```

```python
c0 = ROOT.TCanvas('c0', 'c0', 2400, 700)
ROOT.gStyle.SetPalette(ROOT.kBird)
# ROOT.TColor.InvertPalette()
c0.Divide(3,1)
c0.cd(1)
c0.cd(1).SetLogz()
ROOT.gPad.SetRightMargin(0.13)
ROOT.gPad.SetLeftMargin(0.2)
ROOT.gPad.SetBottomMargin(0.18)
hM_pTcorr_region1.GetXaxis().SetTitle('#sum_{|#eta|<0.2, excluding jet region} p_{T} (GeV)')
hM_pTcorr_region1.GetYaxis().SetTitle('#sum_{jet region} p_{T} (GeV)')
hM_pTcorr_region1.GetXaxis().SetTitleOffset(1.3)
hM_pTcorr_region1.GetYaxis().SetTitleOffset(1.4)
hM_pTcorr_region1.SetContour(2000)
hM_pTcorr_region1.Draw('colz')

profile = hM_pTcorr_region1.ProfileX()
profile.Draw('sameE1')

func = ROOT.TF1('func','pol2',0,1100)
func.FixParameter(0,0)
func.SetLineColor(2)
func.SetLineWidth(2)
profile.Fit(func,'','',0,1100)

ROOT.gROOT.ForceStyle()

ROOT.gPad.Update()
palette = hM_pTcorr_region1.GetListOfFunctions().FindObject('palette')
palette.SetX1NDC(0.88)
palette.SetX2NDC(0.92)
palette.SetY1NDC(0.18)
palette.SetY2NDC(0.95)
ROOT.gPad.Modified()
ROOT.gPad.Update()

c0.cd(2)
ROOT.gPad.SetRightMargin(0.13)
ROOT.gPad.SetLeftMargin(0.2)
ROOT.gPad.SetBottomMargin(0.18)
hM_pTcorr_ratio_region1.GetXaxis().SetTitle('#sum_{|#eta|<0.2, excluding jet region} p_{T} (GeV)')
hM_pTcorr_ratio_region1.GetYaxis().SetTitle('(#sum p_{T}^{jet region})/prediction')
hM_pTcorr_ratio_region1.GetXaxis().SetTitleOffset(1.3)
hM_pTcorr_ratio_region1.GetYaxis().SetTitleOffset(1.5)
hM_pTcorr_ratio_region1.GetXaxis().SetLabelSize(0.05)
hM_pTcorr_ratio_region1.SetContour(2000)
hM_pTcorr_ratio_region1.Draw('colz')

ROOT.gPad.Update()
palette = hM_pTcorr_ratio_region1.GetListOfFunctions().FindObject('palette')
palette.SetX1NDC(0.88)
palette.SetX2NDC(0.92)
palette.SetY1NDC(0.18)
palette.SetY2NDC(0.95)
ROOT.gPad.Modified()
ROOT.gPad.Update()

c0.cd(3)
ROOT.gPad.SetRightMargin(0.05)
ROOT.gPad.SetLeftMargin(0.15)
ROOT.gPad.SetBottomMargin(0.18)
hM_ratio_1D_region1.GetXaxis().SetTitle('(#sum p_{T}^{jet region})/prediction')
hM_ratio_1D_region1.GetYaxis().SetTitle('Entries')
hM_ratio_1D_region1.GetXaxis().SetTitleOffset(1.2)
hM_ratio_1D_region1.GetYaxis().SetTitleOffset(1.3)
hM_ratio_1D_region1.SetLineWidth(2)
hM_ratio_1D_region1.Draw('histE1X0')

Guass = ROOT.TF1('Guass','gaus',0,2)
Guass.SetLineColor(2)
Guass.SetLineWidth(2)
hM_ratio_1D_region1.Fit('Guass')
Guass.Draw('same')

stats = ROOT.TPaveText(0.55, 0.60, 0.92, 0.75, 'NDC')
stats.SetTextFont(42)
stats.AddText('Mean: {0:.3f}#pm{1:.2e}'.format(Guass.GetParameter(1), Guass.GetParError(1)))
stats.AddText('Sigma: {0:.3f}#pm{1:.2e}'.format(Guass.GetParameter(2), Guass.GetParError(2)))
stats.SetBorderSize(0)
stats.SetFillStyle(0)
stats.Draw('same')

c0.Draw()
c0.SaveAs('./plots/Exercise_Tuesday_pTcorr_region1.png')
c0.SaveAs('./plots/Exercise_Tuesday_pTcorr_region1.pdf')
```
